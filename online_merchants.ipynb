Create a function that will check if order-level data is consistent (not different/changing) between rows of the same ORDER_ID.

```python
orders.iloc[23:27, np.r_[0:13, 23:28]]
```

```python
def duplicate(df):
    columns = [col for col in df if not col.startswith("PRODUCT")]
    df['DUPLICATED'] =  df.duplicated(subset=columns, keep='first')
```
